# Area Plot

An area plot also known as an area chart or graph is a type of plot that depicts accumulated totals using numbers or percentages over time. It is based on the line plot and is commonly used when trying to compare two or more quantities.

### Use Case

With so many different types of charts to choose from, there are several points to take into consideration. <br>If someone needs a chart to display changes over a short amount of time, a line chart would be the best bet to use. <br> If they need a chart to show a distribution of data across different subgroups, a bar chart would be more appropriate. <br> When they need a chart to show both change over time and distribution of data across different subgroups, they would use an area chart. 

### Example - Conversion Rate

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'sales': [3, 2, 3, 9, 10, 6],
    'signups': [5, 5, 6, 12, 14, 13],
    'visits': [20, 42, 28, 62, 81, 50],
}, index=pd.date_range(start='2018/01/01', end='2018/07/01',
                       freq='M'))

df1
```

```python
ax = df1.plot.area()
```

```python
# Plot with Matplotlib

import matplotlib.pyplot as plt
import matplotlib as mpl


df1.plot(kind='area',
             stacked=False,
             figsize=(10, 4))

plt.title('Sales Conversion Rate')
plt.ylabel('Rate')
plt.xlabel('Period')

plt.show()
```
