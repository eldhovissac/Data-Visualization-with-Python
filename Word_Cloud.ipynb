```python
!pip install wordcloud
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

import numpy as np
```

```python
with open('Namo Speech.txt', encoding="utf8") as f:
    contents = f.read()
    print(contents)
```

```python
# Creating word_cloud with text as argument in .generate() method
word_cloud = WordCloud(collocations = True, width = 1600, height = 800, contour_color = 'black', background_color = 'grey').generate_from_text(contents)
```

```python
# Display the generated Word Cloud
plt.figure(figsize=(20,10), facecolor='k')
plt.imshow(word_cloud, interpolation='kaiser')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
